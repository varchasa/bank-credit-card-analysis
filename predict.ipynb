```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
logreg = LogisticRegression()
```

```python
df_x = file[['LIMIT_BAL','PAY_0','PAY_2','PAY_3','PAY_4','PAY_5']]
```

```python
df_y=file.CLASS 
```

```python
X_train, X_test, y_train, y_test = train_test_split(df_x, df_y, test_size=0.3, random_state=0) #test and training set
```

```python
logreg.fit(X_train, y_train) #fitting in model
```

```python
y_pred = logreg.predict(X_test) #prediction
```

```python
(logreg.score(X_test, y_test))*100 #accuracy 
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix = confusion_matrix(y_test, y_pred)
print(confusion_matrix) #confusion matrix
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred)) #precision
```
